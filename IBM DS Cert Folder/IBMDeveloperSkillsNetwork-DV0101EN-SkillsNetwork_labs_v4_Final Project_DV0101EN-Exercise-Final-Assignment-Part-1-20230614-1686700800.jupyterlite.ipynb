````python
import pandas as pd
import matplotlib.pyplot as plt

auto_sales_data = df.copy()

# Grouping the data by year and summing up the automobile sales
yearly_sales = auto_sales_data.groupby('Year')['Automobile_Sales'].sum()

# Identifying recession years
recession_years = auto_sales_data[auto_sales_data['Recession'] == 1]['Year'].unique()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(yearly_sales, marker='o', linestyle='-', color='b')

# Adding ticks for each year
plt.xticks(yearly_sales.index, rotation = 90)

# Highlighting recession years
for year in recession_years:
    plt.axvline(x=year, color='r', linestyle='--', alpha=0.7)

# Adding annotations for two recession years (as an example)
if len(recession_years) >= 2:
    plt.annotate('Rec', xy=(recession_years[0], yearly_sales[recession_years[0]]),
                 xytext=(recession_years[0], yearly_sales.max() * 0.8),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    plt.annotate('Rec', xy=(recession_years[1], yearly_sales[recession_years[1]]),
                 xytext=(recession_years[1], yearly_sales.max() * 0.8),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Setting the title and labels
plt.title('Automobile Sales during Recession')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file

auto_sales_data = df.copy()

# Identifying recession years
recession_years = auto_sales_data[auto_sales_data['Recession'] == 1]['Year'].unique()

# Creating a new column to indicate recession and non-recession periods
auto_sales_data['Period'] = auto_sales_data['Year'].apply(lambda x: 'Recession' if x in recession_years else 'Non-Recession')

# Preparing the data for plotting
plot_data = auto_sales_data.groupby(['Year', 'Vehicle_Type', 'Period'])['Automobile_Sales'].sum().reset_index()

# Plotting the trends
plt.figure(figsize=(15, 8))
sns.lineplot(data=plot_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', style='Period', palette='tab10')

# Enhancing the plot
plt.title('Sales Trends per Vehicle Type: Recession vs Non-Recession Periods')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Vehicle Type / Period', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import pandas as pd
import matplotlib.pyplot as plt

# [Include the previous steps of loading data and defining recession periods here]

# Filtering the data to include only non-recession years
non_recession_data_monthly = non_recession_data.groupby('Month').agg({'Automobile_Sales': 'mean', 'Seasonality_Weight': 'mean'}).reset_index()

# Mapping months to numbers for plotting
month_to_num = {month: index for index, month in enumerate(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], 1)}
non_recession_data_monthly['Month_Num'] = non_recession_data_monthly['Month'].map(month_to_num)

# Plotting the bubble chart
plt.figure(figsize=(12, 6))
plt.scatter(non_recession_data_monthly['Month_Num'], non_recession_data_monthly['Automobile_Sales'], 
            s=non_recession_data_monthly['Seasonality_Weight']*1000, alpha=0.5)

# Adding labels and title
plt.title('Seasonality Impact on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)

# Show the plot
plt.show()
````
